# Example 1: Simple Mosaic Plot

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
from itertools import product

# Data to plot
data = {
    ('Male', 'Yes'): 40,
    ('Male', 'No'): 60,
    ('Female', 'Yes'): 50,
    ('Female', 'No'): 50
}

```

```python
# Create the mosaic plot
plt.figure(figsize=(8, 6))
mosaic(data, title='Mosaic Plot of Gender vs. Response')
plt.show()
```
