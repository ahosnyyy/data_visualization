# Example 1: Basic Nested Pie Chart

```python
import matplotlib.pyplot as plt

# Data for outer pie (main categories)
outer_labels = ['Group A', 'Group B', 'Group C']
outer_sizes = [40, 35, 25]
outer_colors = ['skyblue', 'orange', 'lightgreen']

# Data for inner pie (subcategories)
inner_labels = ['A1', 'A2', 'B1', 'B2', 'B3', 'C1', 'C2']
inner_sizes = [20, 20, 15, 10, 10, 15, 10]
inner_colors = ['lightblue', 'lightsteelblue', 'peachpuff', 'moccasin', 'navajowhite', 'palegreen', 'lightgreen']
```

```python
# Create outer pie chart
fig, ax = plt.subplots()
ax.pie(outer_sizes, labels=outer_labels, colors=outer_colors, radius=1.3, wedgeprops=dict(width=0.3, edgecolor='w'))

# Create inner pie chart
ax.pie(inner_sizes, labels=inner_labels, colors=inner_colors, radius=1, wedgeprops=dict(width=0.4, edgecolor='w'), autopct='%1.1f%%')

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.set(aspect="equal")

plt.title('Nested Pie Chart Example')
plt.show()
```
