# Basic Pie Chart
A basic pie chart visualizes the proportions of different categories in a dataset.

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
```

```python
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Basic Pie Chart')
plt.show()
```

# Pie Chart with Exploded Slices
You can highlight a particular slice by "exploding" it out from the pie chart.

```python
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice

# Create a pie chart with exploded slice
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Pie Chart with Exploded Slice')
plt.show()
```

# Pie Chart with Percentages and Labels Outside
Show percentages and labels outside the pie chart.

```python
# Create a pie chart with labels outside
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, pctdistance=0.85)

# Draw a circle at the center to make it look like a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Pie Chart with Labels Outside')
plt.show()
```
