# Basic Violin Plot
A basic violin plot shows the distribution of a dataset along with its density.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data = [np.random.normal(loc=0, scale=1, size=100) for _ in range(3)]
```

```python
# Create a violin plot
plt.figure(figsize=(8, 6))
plt.violinplot(data)
plt.title('Basic Violin Plot')
plt.xlabel('Group')
plt.ylabel('Values')
plt.show()
```

# Violin Plot with Multiple Groups
Comparing distributions across multiple groups.

```python
# Generate sample data
data = [np.random.normal(loc=0, scale=1, size=100),
        np.random.normal(loc=2, scale=1.5, size=100),
        np.random.normal(loc=-2, scale=0.5, size=100)]

# Create a violin plot
plt.figure(figsize=(8, 6))
plt.violinplot(data, showmeans=True, showmedians=True)
plt.title('Violin Plot with Multiple Groups')
plt.xlabel('Group')
plt.ylabel('Values')
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])
plt.show()
```
