# Basic Box Plot
A basic box plot can be used to show the distribution of a dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data = np.random.normal(loc=0, scale=1, size=100)
```

```python
# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.title('Basic Box Plot')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

# Box Plot with Multiple Groups
You can compare distributions across multiple groups.

```python
# Generate sample data
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=1, scale=1.5, size=100)
data3 = np.random.normal(loc=-1, scale=0.5, size=100)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([data1, data2, data3], labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Box Plot with Multiple Groups')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

# Box Plot with Outliers Highlighted
Highlighting outliers can help in understanding data variability.

```python
# Create a box plot with outliers highlighted
plt.figure(figsize=(8, 6))
plt.boxplot(data, flierprops=dict(markerfacecolor='r', marker='o'))
plt.title('Box Plot with Outliers Highlighted')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
