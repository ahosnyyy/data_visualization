# Exercise 1: Box Plot
Objective: Create box plots to analyze the distribution of fares and ages of passengers in the Titanic dataset, comparing distributions across different passenger classes.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
```

```python

# Prepare data for box plots
fare_data = [titanic[titanic['pclass'] == pclass]['fare'] for pclass in titanic['pclass'].unique()]
age_data = [titanic[titanic['pclass'] == pclass]['age'] for pclass in titanic['pclass'].unique()]

# Create a figure with two subplots
plt.figure(figsize=(14, 6))

# Box plot for 'fare' across passenger classes
plt.subplot(1, 2, 1)
plt.boxplot(fare_data, labels=titanic['pclass'].unique())
plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

# Box plot for 'age' across passenger classes
plt.subplot(1, 2, 2)
plt.boxplot(age_data, labels=titanic['pclass'].unique())
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.tight_layout()
plt.show()
```
