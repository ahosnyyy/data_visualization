# Exercise: Violin Plot Analysis of Titanic Data with plt.violinplot
Objective: Create violin plots to analyze the distribution of ages and fares of passengers in the Titanic dataset, comparing distributions across different passenger classes.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
```

```python
# Drop rows with missing values in 'age' and 'fare' columns
titanic = titanic.dropna(subset=['age', 'fare'])

# Prepare data for violin plots
age_data = [titanic[titanic['pclass'] == pclass]['age'].values for pclass in sorted(titanic['pclass'].unique())]
fare_data = [titanic[titanic['pclass'] == pclass]['fare'].values for pclass in sorted(titanic['pclass'].unique())]
pclass_labels = sorted(titanic['pclass'].unique())

# Create a figure with two subplots
plt.figure(figsize=(14, 6))

# Violin plot for 'age' across passenger classes
plt.subplot(1, 2, 1)
plt.violinplot(age_data, showmeans=True, showmedians=True)
plt.title('Violin Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.xticks(ticks=range(1, len(pclass_labels) + 1), labels=pclass_labels)

# Violin plot for 'fare' across passenger classes
plt.subplot(1, 2, 2)
plt.violinplot(fare_data, showmeans=True, showmedians=True)
plt.title('Violin Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.xticks(ticks=range(1, len(pclass_labels) + 1), labels=pclass_labels)

plt.tight_layout()
plt.show()
```
