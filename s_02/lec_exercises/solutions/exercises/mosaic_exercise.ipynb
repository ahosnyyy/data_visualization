# Exercise: Mosaic Plot
Use a mosaic plot to visualize the relationship between gender and survival in the Titanic dataset.

```python
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic

# Load the Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
```

```python
# Prepare data: Create a dictionary with tuple keys for (gender, survival)
data = titanic.groupby(['sex', 'survived']).size().to_dict()

# Create the mosaic plot
plt.figure(figsize=(10, 8))
mosaic(data, title='Mosaic Plot of Gender vs. Survival', gap=0.01)
plt.show()
```
