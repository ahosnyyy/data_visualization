# Exercise 1: ECDF
Question: Plot the ECDF for the ages of passengers in the Titanic dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
```

```python
# Drop missing values for the 'age' column
df_age = titanic['age'].dropna()

# Calculate ECDF
sorted_age = np.sort(df_age)
ecdf = np.arange(1, len(sorted_age) + 1) / len(sorted_age)

# Plot ECDF
plt.figure(figsize=(8, 6))
plt.step(sorted_age, ecdf, where='post', label='ECDF of Age')
plt.xlabel('Age')
plt.ylabel('ECDF')
plt.title('ECDF of Age in Titanic Dataset')
plt.grid(True)
plt.legend()
plt.show()
```

# Exercise 2: Q-Q
Question: Plot the Q-Q plot for the ages of passengers in the Titanic dataset.

```python
# Drop missing values for the 'age' column
df_age = titanic['age'].dropna()

# Create Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(df_age, dist="norm", plot=plt)
plt.title('Q-Q Plot of Age in Titanic Dataset')
plt.show()
```
