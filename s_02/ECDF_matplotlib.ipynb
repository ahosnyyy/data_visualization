## ECDF

```python
import matplotlib.pyplot as plt
import numpy as np

# Hypothetical exam scores for 50 students
scores = np.random.randint(0, 101, 50)
```

```python
# Sort the scores in ascending order
sorted_scores = np.sort(scores)

# Calculate the ranks (cumulative distribution)
ranks = np.arange(1, len(sorted_scores) + 1) / len(sorted_scores)

# Plotting the ECDF
plt.figure(figsize=(10, 6))
plt.step(sorted_scores, ranks, where='post', color='blue', label='ECDF')

# Customize the Plot
plt.xlabel('Exam Scores')
plt.ylabel('Cumulative Probability')
plt.title('Empirical Cumulative Distribution Function (ECDF) of Exam Scores')
plt.grid(True)
plt.legend()
plt.show()
```

## Highly Skewed Data

```python
from scipy.stats import gaussian_kde

# Step 1: Simulate a Highly Skewed Dataset
np.random.seed(42)
data = np.random.lognormal(mean=10, sigma=1, size=1000)

# Apply a Log Transformation
log_data = np.log(data)

# Step 2: Plot the Density Plot
plt.figure(figsize=(14, 6))

# Density Plot for Original Data
plt.subplot(2, 2, 1)
density_original = gaussian_kde(data)
x_vals_original = np.linspace(min(data), max(data), 1000)
plt.plot(x_vals_original, density_original(x_vals_original), color='skyblue')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Plot of Original Skewed Data')
plt.grid(True)

# Density Plot for Log-Transformed Data
plt.subplot(2, 2, 2)
density_log = gaussian_kde(log_data)
x_vals_log = np.linspace(min(log_data), max(log_data), 1000)
plt.plot(x_vals_log, density_log(x_vals_log), color='lightcoral')
plt.xlabel('Log(Value)')
plt.ylabel('Density')
plt.title('Density Plot of Log-Transformed Data')
plt.grid(True)

# Step 3: Plot the ECDF

# ECDF for Original Data
plt.subplot(2, 2, 3)
sorted_data = np.sort(data)
ranks = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
plt.step(sorted_data, ranks, where='post', color='skyblue')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('ECDF of Original Skewed Data')
plt.grid(True)

# ECDF for Log-Transformed Data
plt.subplot(2, 2, 4)
sorted_log_data = np.sort(log_data)
ranks_log = np.arange(1, len(sorted_log_data) + 1) / len(sorted_log_data)
plt.step(sorted_log_data, ranks_log, where='post', color='lightcoral')
plt.xlabel('Log(Value)')
plt.ylabel('Cumulative Probability')
plt.title('ECDF of Log-Transformed Data')
plt.grid(True)

plt.tight_layout()
plt.show()
```
