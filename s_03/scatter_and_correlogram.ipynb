# Scatterplots
Scatterplots are great for visualizing the relationship between two continuous variables.

## 0. Basic Example

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', marker='o', edgecolor='k', s=100)  # 's' is the marker size

# Add titles and labels
plt.title('Scatterplot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.grid(True)
plt.show()
```

## 1. Scatterplot with Different Marker Sizes and Colors


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100
sizes = np.random.rand(50) * 1000  # Bubble sizes
colors = np.random.rand(50)  # Bubble colors

plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.6, edgecolors='w', linewidth=0.5)

# Add color bar
plt.colorbar(scatter, label='Color Scale')

# Add titles and labels
plt.title('Scatterplot with Different Marker Sizes and Colors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.grid(True)
plt.show()
```

## 2. Scatterplot with Regression Line

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = 2.5 * x + np.random.normal(0, 1, 100)

plt.figure(figsize=(10, 6))
sns.regplot(x=x, y=y, scatter_kws={'color': 'blue', 's': 50}, line_kws={'color': 'red', 'lw': 2})

# Add titles and labels
plt.title('Scatterplot with Regression Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.grid(True)
plt.show()
```

## 3. Scatterplot with Annotations

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
labels = ['A', 'B', 'C', 'D', 'E']

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='green', marker='o')

# Annotate points
for i, label in enumerate(labels):
    plt.annotate(label, (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Add titles and labels
plt.title('Scatterplot with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.grid(True)
plt.show()
```

## 4. 3D Scatterplot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
z = np.random.rand(100) * 100

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='purple', marker='o')

# Add titles and labels
ax.set_title('3D Scatterplot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Show plot
plt.show()
```

## 5. Scatterplot with Different Marker Styles


```python
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, c='blue', marker='x', label='Cross Marker')
plt.scatter(x + 1, y + 1, c='red', marker='^', label='Triangle Marker')
plt.scatter(x + 2, y - 1, c='green', marker='o', label='Circle Marker')

# Add titles and labels
plt.title('Scatterplot with Different Marker Styles')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

# Correlogram 

A correlogram is a graphical representation of the correlation matrix of a dataset. It is often used to visualize the correlation between different variables.

## Example 1: Basic Correlogram Using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
np.random.seed(0)
data = pd.DataFrame(np.random.rand(100, 5), columns=['A', 'B', 'C', 'D', 'E'])

# Compute the correlation matrix
corr_matrix = data.corr()

# Create the correlogram
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0, linewidths=0.5)

# Add titles
plt.title('Correlogram')

# Show plot
plt.show()
```

## Example 2: Correlogram with Different Color Maps and Customization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
np.random.seed(0)
data = pd.DataFrame(np.random.rand(100, 7), columns=['A', 'B', 'C', 'D', 'E', 'F', 'G'])

# Compute the correlation matrix
corr_matrix = data.corr()

# Create the correlogram with custom color map
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt='.1f', vmin=-1, vmax=1, center=0, linewidths=0.3, linecolor='gray')

# Add titles and customize labels
plt.title('Correlogram with Custom Color Map and Labels')
plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Show plot
plt.show()
```

## Example 3: Correlogram with Hierarchical Clustering
Hierarchical clustering can be added to reorder the heatmap for better visualization of patterns.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
np.random.seed(0)
data = pd.DataFrame(np.random.rand(100, 5), columns=['A', 'B', 'C', 'D', 'E'])

# Compute the correlation matrix
corr_matrix = data.corr()

# Create a cluster map with hierarchical clustering
plt.figure(figsize=(10, 8))
sns.clustermap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0, linewidths=0.5)

# Add title
plt.title('Correlogram with Hierarchical Clustering')

# Show plot
plt.show()
```
