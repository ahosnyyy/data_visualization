```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sample DataFrame
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
})
```

# 1. Basic Dot Plot

```python
# Dot Plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Category'], df['Values'], s=100, color='blue', alpha=0.6, edgecolor='k')
plt.title('Basic Dot Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

# 2. Dot Plot with Multiple Categories

```python
# Sample DataFrame with multiple categories
df2 = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'A', 'B', 'C', 'D'],
    'Subcategory': ['X', 'X', 'X', 'X', 'Y', 'Y', 'Y', 'Y'],
    'Values': [10, 20, 15, 25, 12, 22, 14, 27]
})

# Dot Plot with multiple categories
plt.figure(figsize=(8, 5))
colors = {'X':'blue', 'Y':'green'}
plt.scatter(df2['Category'], df2['Values'], s=100, c=df2['Subcategory'].map(colors), alpha=0.6, edgecolor='k')
plt.title('Dot Plot with Multiple Categories')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='X'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='Y')],
           title='Subcategory')
plt.show()
```

# 3. Dot Plot with Different Sizes

```python
# Sample DataFrame with different sizes
df3 = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25],
    'Size': [30, 60, 45, 90]  # Size for the dots
})

# Dot Plot with different sizes
plt.figure(figsize=(8, 5))
plt.scatter(df3['Category'], df3['Values'], s=df3['Size'], color='blue', alpha=0.6, edgecolor='k')
plt.title('Dot Plot with Different Sizes')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

# 4. Dot Plot with Custom Color

```python
# Dot Plot with custom color
plt.figure(figsize=(8, 5))
plt.scatter(df['Category'], df['Values'], s=100, color='purple', alpha=0.6, edgecolor='k')
plt.title('Dot Plot with Custom Color')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
