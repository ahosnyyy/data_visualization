```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate sample data
data = np.random.randn(1000)
```

# 1. Basic Histogram

```python
# Basic Histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# 2. Histogram with Density

```python
# Histogram with density
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Histogram with Density')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

# 3. Histogram with Multiple Datasets

```python
# Generate sample data
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2

# Histogram with multiple datasets
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=30, alpha=0.5, label='Data 1', color='blue', edgecolor='black')
plt.hist(data2, bins=30, alpha=0.5, label='Data 2', color='red', edgecolor='black')
plt.title('Histogram with Multiple Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

# 4. Cumulative Histogram

```python
# Cumulative Histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, cumulative=True, color='green', edgecolor='black')
plt.title('Cumulative Histogram')
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')
plt.grid(True)
plt.show()
```

# 5. Histogram with Normal Distribution Overlay

```python
import scipy.stats as stats

# Histogram with normal distribution overlay
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, color='orange', edgecolor='black', alpha=0.6)

# Plot normal distribution curve
mu, std = np.mean(data), np.std(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Histogram with Normal Distribution Overlay')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
