# Exercise 1: Density Plot of Passenger Ages
Question: Create a density plot showing the distribution of passenger ages on the Titanic using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
import pandas as pd

# Load the Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')

# Prepare the Data
ages = titanic['age'].dropna()

# Compute the KDE
kde = gaussian_kde(ages, bw_method=0.5)  # Adjust bandwidth as needed
x = np.linspace(ages.min(), ages.max(), 1000)
density = kde(x)

# Create the Density Plot
plt.figure(figsize=(8, 5))
plt.plot(x, density, color='skyblue', alpha=0.7)

# Customize the Plot
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Density Plot of Passenger Ages on the Titanic')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Exercise 2: Density Plot of Passenger Ages by Class
Question: Create density plots to compare the distribution of passenger ages across different classes (Pclass) using Matplotlib.

```python
# Prepare the Data
classes = sorted(titanic['pclass'].unique())

# Create the Density Plot
plt.figure(figsize=(12, 8))
for class_ in classes:
    ages = titanic[titanic['pclass'] == class_]['age'].dropna()
    kde = gaussian_kde(ages, bw_method=0.3)
    x = np.linspace(ages.min(), ages.max(), 1000)
    density = kde(x)
    plt.plot(x, density, label=f'Class {class_}', alpha=0.7)

# Customize the Plot
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Density Plot of Passenger Ages by Class on the Titanic')
plt.legend(title='Class')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
