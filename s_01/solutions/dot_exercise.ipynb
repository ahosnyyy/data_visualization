# Exercise: Dot Plot of Survival Status Counts by Class
Question: Create a dot plot showing the number of survivors and non-survivors within each class.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')

# Prepare the Data
survival_class_counts = titanic.groupby(['pclass', 'survived']).size().unstack().fillna(0)

# Create the Dot Plot
plt.figure(figsize=(10, 6))
colors = {0: 'red', 1: 'green'}
for survived in survival_class_counts.columns:
    plt.plot(survival_class_counts[survived], survival_class_counts.index, 'o', 
             color=colors[survived], label='Survived' if survived == 1 else 'Not Survived', alpha=0.7)

# Customize the Plot
plt.xlabel('Number of Passengers')
plt.ylabel('Passenger Class')
plt.title('Dot Plot of Survival Status Counts by Class on the Titanic')

# Set Passenger Class as Discrete Values
plt.yticks(ticks=[1, 2, 3], labels=['Class 1', 'Class 2', 'Class 3'])

plt.legend(title='Survival Status')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
