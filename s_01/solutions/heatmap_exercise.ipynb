# Exercise 1: Heatmap of Survival Status by Passenger Class and Gender
Question: Create a heatmap showing the count of survivors and non-survivors across different passenger classes (Pclass) and genders (sex).

HINT: Use pandas `pivot_table`

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')


# Prepare the Data
pivot_table = titanic.pivot_table(index='pclass', columns='sex', values='survived', aggfunc='sum').fillna(0)

# Create the Heatmap
plt.figure(figsize=(10, 6))
plt.imshow(pivot_table, cmap='coolwarm', interpolation='nearest')

# Add Colorbar
plt.colorbar(label='Number of Survivors')

# Customize the Plot
plt.xlabel('Gender')
plt.ylabel('Passenger Class')
plt.title('Heatmap of Survivors by Passenger Class and Gender on the Titanic')

# Set axis ticks and labels
plt.xticks(ticks=np.arange(len(pivot_table.columns)), labels=pivot_table.columns)
plt.yticks(ticks=np.arange(len(pivot_table.index)), labels=pivot_table.index)

plt.show()
```

# Exercise 2: Heatmap of Average Fare by Embarked Location and Passenger Class
Question: Create a heatmap showing the average fare paid by passengers based on their embarkation location and passenger class.

```python
# Prepare the Data
pivot_table = titanic.pivot_table(index='pclass', columns='embarked', values='fare', aggfunc='mean').fillna(0)

# Create the Heatmap
plt.figure(figsize=(10, 6))
plt.imshow(pivot_table, cmap='viridis', interpolation='nearest')

# Add Colorbar
plt.colorbar(label='Average Fare')

# Customize the Plot
plt.xlabel('Embarked Location')
plt.ylabel('Passenger Class')
plt.title('Heatmap of Average Fare by Embarked Location and Passenger Class on the Titanic')

# Set axis ticks and labels
plt.xticks(ticks=np.arange(len(pivot_table.columns)), labels=pivot_table.columns)
plt.yticks(ticks=np.arange(len(pivot_table.index)), labels=pivot_table.index)

plt.show()
```
