# Exercise 1: Grouped Bar Plot
Question: Create a grouped bar plot to compare the number of passengers by class and gender in the Titanic dataset.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')

# Prepare the Data
class_gender_counts = titanic.groupby(['pclass', 'sex']).size().unstack()

# Set up bar positions
x = np.arange(len(class_gender_counts.index))
width = 0.35

# Create the Grouped Bar Plot
plt.figure(figsize=(10, 6))
bars1 = plt.bar(x - width/2, class_gender_counts['male'], width, label='Male', color='skyblue')
bars2 = plt.bar(x + width/2, class_gender_counts['female'], width, label='Female', color='lightcoral')

# Customize the Plot
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Class and Gender')
plt.xticks(x, class_gender_counts.index)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value annotations on top of the bars
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, f'{yval}', ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, f'{yval}', ha='center', va='bottom')

plt.show()
```

# Exercise 2: Horizontal Bar Plot
Question: Create a horizontal bar plot to display the number of passengers by class in the Titanic dataset.

```python
# Prepare the Data
class_counts = titanic['pclass'].value_counts().sort_index()

# Create the Horizontal Bar Plot
plt.figure(figsize=(10, 6))
plt.barh(class_counts.index.astype(str), class_counts.values, color='skyblue')

# Customize the Plot
plt.xlabel('Number of Passengers')
plt.ylabel('Class')
plt.title('Number of Passengers by Class on the Titanic')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# Exercise 3: Stacked Bar Plot
Question: Create a stacked bar plot to show the number of survivors and non-survivors by class.

```python
# Prepare the Data
class_survival_counts = titanic.groupby(['pclass', 'survived']).size().unstack()
class_survival_counts = class_survival_counts.fillna(0)

# Create the Stacked Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(class_survival_counts.index.astype(str), class_survival_counts[0], color='lightcoral', label='Not Survived')
plt.bar(class_survival_counts.index.astype(str), class_survival_counts[1], bottom=class_survival_counts[0], color='skyblue', label='Survived')

# Customize the Plot
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Class and Survival Status')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
