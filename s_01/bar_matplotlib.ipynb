```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sample DataFrame
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
})
```

# 1. Vertical Bar Plot
The most common type of bar plot, where bars are vertical.

```python
# Vertical Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Vertical Bar Plot')
plt.show()
```

# 2. Horizontal Bar Plot
Bar plot where bars are horizontal.

```python
# Horizontal Bar Plot
plt.figure(figsize=(8, 5))
plt.barh(df['Category'], df['Values'], color='lightgreen')
plt.xlabel('Values')
plt.ylabel('Category')
plt.title('Horizontal Bar Plot')
plt.show()
```

# 3. Stacked Bar Plot
Shows multiple datasets stacked on top of each other.

```python
df2 = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values1': [10, 20, 15, 25],
    'Values2': [5, 10, 10, 15]
})

# Stacked Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(df2['Category'], df2['Values1'], color='skyblue', label='Values1')
plt.bar(df2['Category'], df2['Values2'], bottom=df2['Values1'], color='lightcoral', label='Values2')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.legend()
plt.show()
```

```python
df2 = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values1': [10, 20, 15, 25],
    'Values2': [5, 10, 10, 15]
})

# Stacked Bar Plot
plt.figure(figsize=(8, 5))
bars1 = plt.bar(df2['Category'], df2['Values1'], color='skyblue', label='Values1')
bars2 = plt.bar(df2['Category'], df2['Values2'], bottom=df2['Values1'], color='lightcoral', label='Values2')

# Add numbers on top of the bars
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height - 3, str(int(height)),
             ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height() + bar.get_y()  # height + bottom (stacked part)
    plt.text(bar.get_x() + bar.get_width() / 2, height - 3, str(int(height - bar.get_y())),
             ha='center', va='bottom')

plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.legend()
plt.show()
```

# 4. Grouped Bar Plot
Displays bars for different groups side by side.

```python
import numpy as np

# Grouped Bar Plot
x = np.arange(len(df2['Category']))
width = 0.35  # width of bars

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, df2['Values1'], width, color='skyblue', label='Values1')
plt.bar(x + width/2, df2['Values2'], width, color='lightcoral', label='Values2')

plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Grouped Bar Plot')
plt.xticks(x, df2['Category'])
plt.legend()
plt.show()
```

# 5. Error Bar Plot
Includes error bars to show variability or uncertainty.

```python
# Sample DataFrame with errors
df3 = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25],
    'Errors': [1, 2, 1.5, 2.5]
})

# Error Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(df3['Category'], df3['Values'], yerr=df3['Errors'], color='skyblue', capsize=5)
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Error Bar Plot')
plt.show()
```

# 6. Cumulative Bar Plot
Shows cumulative totals for each category.

```python
df4 = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [5, 15, 10, 20]
})

# Cumulative Bar Plot
df4['Cumulative'] = df4['Values'].cumsum()

plt.figure(figsize=(8, 5))
plt.bar(df4['Category'], df4['Cumulative'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Cumulative Values')
plt.title('Cumulative Bar Plot')
plt.show()
```

# 7. Bar Plot with Multiple Axes
Combines different types of bar plots in a single figure with multiple subplots.

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Vertical Bar Plot
axs[0].bar(df['Category'], df['Values'], color='skyblue')
axs[0].set_title('Vertical Bar Plot')
axs[0].set_xlabel('Category')
axs[0].set_ylabel('Values')

# Horizontal Bar Plot
axs[1].barh(df['Category'], df['Values'], color='lightgreen')
axs[1].set_title('Horizontal Bar Plot')
axs[1].set_xlabel('Values')
axs[1].set_ylabel('Category')

plt.tight_layout()
plt.show()
```
