```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create sample data
data = np.random.rand(10, 12)
```

# 1. Basic Heatmap

```python
# Basic Heatmap
plt.figure(figsize=(10, 8))
plt.imshow(data, cmap='viridis', aspect='auto')
plt.colorbar(label='Intensity')
plt.title('Basic Heatmap')
plt.show()
```

# 2. Heatmap with Custom Color Map

```python
# Heatmap with custom color map
plt.figure(figsize=(10, 8))
plt.imshow(data, cmap='coolwarm', aspect='auto')
plt.colorbar(label='Intensity')
plt.title('Heatmap with Custom Color Map')
plt.show()
```

# 3. Heatmap with Annotations

```python
# Heatmap with annotations
plt.figure(figsize=(10, 8))
plt.imshow(data, cmap='YlGnBu', aspect='auto')

# Adding annotations
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        plt.text(j, i, f'{data[i, j]:.2f}', ha='center', va='center', color='black')

plt.colorbar(label='Intensity')
plt.title('Heatmap with Annotations')
plt.show()
```

# 4. Heatmap with Masking

```python
# Create mask
mask = np.triu(np.ones_like(data, dtype=bool))

# Apply mask to the data
masked_data = np.ma.masked_where(mask, data)

# Heatmap with masking
plt.figure(figsize=(10, 8))
plt.imshow(masked_data, cmap='BuPu', aspect='auto')
plt.colorbar(label='Intensity')
plt.title('Heatmap with Masking')
plt.show()
```
