# Exercise 1: Histogram of Passenger Ages
Question: Create a histogram showing the distribution of passenger ages on the Titanic. Make sure to handle missing values appropriately.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')

# Prepare the Data
ages = titanic['age'].dropna()

# Create the Histogram
plt.figure(figsize=(8, 5))
plt.hist(ages, bins=20, color='skyblue', edgecolor='black', alpha=0.7)

# Customize the Plot
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Histogram of Passenger Ages on the Titanic')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Exercise 2: Histogram of Passenger Ages by Survival Status
Question: Create histograms to compare the distribution of passenger ages for those who survived versus those who did not survive. Use subplots to show both distributions side by side.

```python
# Prepare the Data
survived_ages = titanic[titanic['survived'] == 1]['age'].dropna()
not_survived_ages = titanic[titanic['survived'] == 0]['age'].dropna()

# Create the Histograms
plt.figure(figsize=(12, 5))

# Histogram for Survived
plt.subplot(1, 2, 1)
plt.hist(survived_ages, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Ages of Survived Passengers')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Histogram for Not Survived
plt.subplot(1, 2, 2)
plt.hist(not_survived_ages, bins=20, color='lightcoral', edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Ages of Not Survived Passengers')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
