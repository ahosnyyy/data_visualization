```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
# Generate sample data
data = np.random.randn(1000)
```

# 1. Basic Density Plot

```python
# Compute the KDE
kde = gaussian_kde(data, bw_method='scott')  # 'scott' or 'silverman' can be used
x = np.linspace(min(data), max(data), 1000)
kde_values = kde(x)

# Plot the KDE
plt.figure(figsize=(10, 6))
plt.plot(x, kde_values, color='blue', label='Density Estimate')
plt.fill_between(x, kde_values, color='blue', alpha=0.5)
plt.title('Basic Density Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```

# 2. Density Plot with Multiple Datasets

```python
# Generate sample data
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2

# Compute KDE for both datasets
kde1 = gaussian_kde(data1, bw_method='scott')
kde2 = gaussian_kde(data2, bw_method='scott')
x = np.linspace(min(min(data1), min(data2)), max(max(data1), max(data2)), 1000)
kde_values1 = kde1(x)
kde_values2 = kde2(x)

# Plot the KDEs
plt.figure(figsize=(10, 6))
plt.plot(x, kde_values1, color='blue', label='Data 1 Density')
plt.fill_between(x, kde_values1, color='blue', alpha=0.5)
plt.plot(x, kde_values2, color='red', label='Data 2 Density')
plt.fill_between(x, kde_values2, color='red', alpha=0.5)
plt.title('Density Plot with Multiple Datasets')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```

# 3. Density Plot with Different Bandwidths

```python
# Compute KDE with different bandwidths
kde_bw1 = gaussian_kde(data, bw_method=0.5)
kde_bw2 = gaussian_kde(data, bw_method=1.0)
kde_bw3 = gaussian_kde(data, bw_method=2.0)
x = np.linspace(min(data), max(data), 1000)
kde_values_bw1 = kde_bw1(x)
kde_values_bw2 = kde_bw2(x)
kde_values_bw3 = kde_bw3(x)

# Plot the KDEs
plt.figure(figsize=(10, 6))
plt.plot(x, kde_values_bw1, color='blue', label='Bandwidth 0.5')
plt.plot(x, kde_values_bw2, color='green', label='Bandwidth 1.0')
plt.plot(x, kde_values_bw3, color='red', label='Bandwidth 2.0')
plt.fill_between(x, kde_values_bw1, color='blue', alpha=0.3)
plt.fill_between(x, kde_values_bw2, color='green', alpha=0.3)
plt.fill_between(x, kde_values_bw3, color='red', alpha=0.3)
plt.title('Density Plot with Different Bandwidths')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```

# 4. Density Plot with Rug Plot

```python
# Compute KDE
kde = gaussian_kde(data, bw_method='scott')
x = np.linspace(min(data), max(data), 1000)
kde_values = kde(x)

# Plot the KDE with rug plot
plt.figure(figsize=(10, 6))
plt.plot(x, kde_values, color='blue', label='Density Estimate')
plt.fill_between(x, kde_values, color='blue', alpha=0.5)

# Adding rug plot
plt.scatter(data, np.zeros_like(data), color='black', s=10, label='Rug Plot')

plt.title('Density Plot with Rug Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```
